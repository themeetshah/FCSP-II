```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
f = open('indeed jobs.html', 'r')
doc = BeautifulSoup(f.read(), 'html.parser')

job_names_tags = doc.find_all('h2', class_='jobTitle css-1psdjh5 eu4oa1w0')
job_companies_tags = doc.find_all('div', class_='css-1afmp4o e37uo190')
job_locs_tags = doc.find_all('div', class_='css-1restlb eu4oa1w0')
job_details_tags = doc.find_all('div', class_='jobMetaDataGroup css-qspwa8 eu4oa1w0')

job_names=[]
job_companies=[]
job_locs=[]
job_details=[]
for i in range(len(job_names_tags)):
    job_names.append(job_names_tags[i].text)
    job_companies.append(job_companies_tags[i].text)
    job_locs.append(job_locs_tags[i].text)
    job_details.append(job_details_tags[i].text)

df = pd.DataFrame({'name-of-the-job':job_names, 'name-of-the-companies':job_companies, 'job-location':job_locs, 'job-details':job_details})
df
```
