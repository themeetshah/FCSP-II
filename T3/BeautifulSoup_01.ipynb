```python
import pandas as pd
response = requests.get("https://www.politifact.com/factchecks/list/")
content = response.text

soup = BeautifulSoup(content, 'html.parser')
source = soup.find_all('a',class_="m-statement__name")
statement = soup.find_all('div',class_="m-statement__quote")
date = soup.find_all('footer',class_="m-statement__footer")
tag = soup.find_all('img',class_="c-image__original", alt=True)

d = {'source':[],'statement':[],'date':[],'tag':[]}
for i in range(len(source)):
    d['source'].append(source[i].text.strip())
    d['statement'].append(statement[i].text.strip())
    d['date'].append(date[i].text.split("•")[1].strip())
    d['tag'].append(tag[i]['alt'].strip())
df= pd.DataFrame(d)
df
```

```python
response = requests.get("https://www.cricbuzz.com/cricket-series/9237/indian-premier-league-2025/points-table")
content = response.text

soup = BeautifulSoup(content, 'html.parser')
table = soup.find('div',class_='cb-col cb-col-84')
s=soup.find_all('td',class_='cb-srs-pnts-td')

mat = s[0::7]
won = s[1::7]
lost = s[2::7]
tied = s[3::7]
NR = s[4::7]
pts = s[5::7]
nrr = s[6::7]
d = {'match':[],'won':[],'lost':[],'tied':[],'NR':[],'pts':[],'nrr':[]}
for i in range(len(mat)):
    d['match'].append(mat[i].text.strip())
    d['won'].append(won[i].text.strip())
    d['lost'].append(lost[i].text.strip())
    d['tied'].append(tied[i].text.strip())
    d['NR'].append(NR[i].text.strip())
    d['pts'].append(pts[i].text.strip())
    d['nrr'].append(nrr[i].text.strip())
df= pd.DataFrame(d)
df
```

```python
f = open('newflipkart.html')
content = f.read()
soup = BeautifulSoup(content, 'html.parser')

name_tags = soup.find_all('div', class_='KzDlHZ')
names = []
memory = []
color = []
# name = Name (color, memory)
for tag in name_tags:
    full_text = tag.text.strip()

    if '(' in full_text and ')' in full_text:
        info = full_text.split('(')[1].split(')')[0]
        names.append(full_text.split('(')[0])
        parts = [p.strip() for p in info.split(',')]
        if len(parts) == 2:
            color.append(parts[0])
            memory.append(parts[1])

star_tags = soup.find_all('div', class_='XQDdHH')
stars = []
for tag in star_tags:
    full_text = tag.text.strip()
    stars.append(full_text)

review_tags = soup.find_all('span', class_='Wphh3N')
reviews = []
for tag in review_tags:
    full_text = tag.text.strip()
    reviews.append(full_text.split()[3])

old_tags = soup.find_all('div', class_='yRaY8j ZYYwLA')
old_prices = []
for tag in old_tags:
    full_text = tag.text.strip()
    old_prices.append(full_text)

new_tags = soup.find_all('div', class_='Nx9bqj _4b5DiR')
new_prices = []
for tag in new_tags:
    full_text = tag.text.strip()
    new_prices.append(full_text)

differences = []
for i in range(len(old_prices)):
    old_prices[i] = old_prices[i].replace('₹', '')
    old_prices[i] = old_prices[i].replace(',', '')
    old_prices[i] = int(old_prices[i])
    new_prices[i] = new_prices[i].replace('₹', '')
    new_prices[i] = new_prices[i].replace(',', '')
    new_prices[i] = int(new_prices[i])
    differences.append(old_prices[i]-new_prices[i])

disc_tags = soup.find_all('div', class_='UkUFwK')
discounts = []
for tag in disc_tags:
    full_text = tag.text.strip()
    discounts.append(full_text.split()[0])

df = pd.DataFrame(
    {
        'Name': names,
        'Color': color,
        'Memory': memory,
        'Stars':stars,
        'Reviews':reviews,
        'Old Price':old_prices,
        'New Price':new_prices,
        'Difference':differences,
        'Discount':discounts
    }
)
df
```
