```python
market = pd.read_csv('supermarket.csv')
df = pd.DataFrame(market)
```

```python
# 1.
df.head(10)
```

```python
df.isna().sum()
```

```python
#2. 
df.fillna(df.mean(numeric_only=True), inplace=True)
df.isna().sum()
```

```python
df.fillna(df.mode().loc[0], inplace=True)
df.isna().sum()
```

```python
df['branch'].unique()
```

```python
# 3.
print('Revenue from A:',df[df['branch']=='A']['total'].sum())
print('Revenue from B:',df[df['branch']=='B']['total'].sum())
print('Revenue from C:',df[df['branch']=='C']['total'].sum())
```

```python
list(df['product line'].unique())
```

```python
# 4.
pl = list(df['product line'].unique())
pl_dict = {}
for i in pl:
    pl_dict[i]=len(df[df['product line']==i])
most = (sorted(pl_dict.items(), key=lambda item:item[1], reverse=True))[0]
most
```

```python
# 5.
print('Avg. Rating for A:',df[df['branch']=='A']['rating'].mean())
print('Avg. Rating for B:',df[df['branch']=='B']['rating'].mean())
print('Avg. Rating for C:',df[df['branch']=='C']['rating'].mean())
```

```python
# 6.
pay = list(df['payment'].unique())
pay_dict = {}
for i in pay:
    pay_dict[i]=len(df[df['payment']==i])
most = (sorted(pay_dict.items(), key=lambda item:item[1], reverse=True))[0]
most
```

```python
# 7.
len(df[df['customer type']=='Member'])
```

```python
# 8.
pd.crosstab(df['customer type'],df['payment'])
```

```python
# 9.
print('Weekends:',df.loc[(df['day_of_week']=='Saturday') | (df['day_of_week']=='Sunday')]['total'].sum())
print('Weekdays:',df.loc[(df['day_of_week']!='Saturday') & (df['day_of_week']!='Sunday')]['total'].sum())
```
