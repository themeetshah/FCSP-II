```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([[1,2],[4,8],[9,11]], index = ['A','B','C'], columns=['X','Y'])
print("loc['C']:")
print(df.loc['C'])
print()

print("loc['B':'C']:")
print(df.loc['B':'C'])
print()

print("loc[:,'X']:")
print(df.loc[:,'X'])
print()
df
```

#### -- passing a list as df.loc[['C']] will give dataframe else if df.loc['C'] will give data series

```python
df.loc[['C']] # list pass kari etle dataframe
```

```python
df.loc['C']  # list pass nthi kari etle data series
```

```python
df.loc[['A','C'],'X']
```

```python
df.loc[df['X']>5]  #2D ma ans aavse always even if ek j value condition satisfy kare
```

```python
df.loc[df['X']>3]  #evi rows jema X>3 hoy
```

```python
df.loc[df['X']>3,'Y']  #evi Y ni values je rows jema X>3 hoy
```

```python
df.loc[lambda df : df['Y'] == 8]
```

```python
df.loc[(df['X']>3) & (df['Y']<10)]  # bitwise and 
```

```python
df.loc[(df['X']>3) | (df['Y']<10)]  # bitwise or
```

```python
df.loc[['B','C'],'Y'] = 50  # update those values
df
```

```python
df.loc['A']=10
df
```

```python
df.loc[df['X']>5] = 0  # if condition satisfy thai toh aakhi row ni values change
df
```

```python
df = pd.DataFrame({'A':[1,2,np.nan,4,np.nan],
                   'B':[np.nan,6,7,np.nan,np.nan],
                   'C':[10,4,np.nan,np.nan,14]})
df
```

```python
df.dropna(thresh=2)  # 2 or 2 thi vadhare non-null values je row ma hase e rows return karse
```

```python
df.dropna(thresh=3)
```

```python
df.dropna(subset=['B','C'])  # B or C ma null values je row ma hase e row drop thase
```

```python
df = pd.DataFrame({'A':[1,1,2,3,4,4],
                   'B':[5,5,6,7,8,8]})
df.drop_duplicates()  # first duplicate raakhvani
# aakhi row duplicate rehvi joie
```

```python
df.drop_duplicates(keep='last')  # last duplicate raakhvani
```

```python
df = pd.DataFrame({'A':[1,1,2,3,4,4],
                   'B':[5,5,6,7,8,8],
                   'C':['X','X','Y','X','W','Z']})
df.drop_duplicates()
```
