```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Area Plot

```python
x = range(1,6)
y = [1,4,6,8,4]
plt.fill_between(x,y)
plt.show()
```

```python
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020],
    'Sales': [100, 120, 150, 180, 200, 220],
    'Expenses': [70, 80, 90, 100, 110, 120]
}
df = pd.DataFrame(data)
# 1st x-axis and rest y-axis
# area plot jevu j che
plt.stackplot(df['Year'], df['Sales'], df['Expenses'], labels = ['Sales', 'Expenses'])
plt.legend(loc='upper left')
plt.show()
```

```python
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020],
    'Sales': [100, 120, 150, 170, 200, 220],
    'Expenses': [70, 80, 90, 100, 110, 120]
}
df = pd.DataFrame(data)
# 1st x-axis and rest y-axis
plt.fill_between(df['Year'], df['Sales'], df['Expenses'])
plt.show()
```

```python
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020],
    'Sales': [100, 60, 150, 170, 200, 220],
    'Expenses': [70, 80, 90, 100, 110, 120]
}
df = pd.DataFrame(data)
# 1st x-axis and rest y-axis
plt.fill_between(df['Year'], df['Sales'], df['Expenses'])
plt.show()
```

```python
months = np.arange(1, 13)
sales = np.array([10, 15, 20, 30, 20, 50, 30, 70, 80, 40, 100, 10])
expenses = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60])

plt.plot(months, sales, color='green', label='sales')
plt.plot(months, expenses, color='red', label='expenses')
plt.fill_between(months, sales, expenses, where=sales>expenses, alpha=0.25, color='green', interpolate=True)
plt.fill_between(months, sales, expenses, where=sales<expenses, alpha=0.25, color='red', interpolate=True)
plt.legend()
plt.show()
```

## Scatter Plot

```python
x1 = 5 * np.random.rand(50)
x2 = 5 * np.random.rand(50) + 25
x3 = 30 * np.random.rand(25)
x = np.concatenate([x1,x2,x3])

y1 = 5 * np.random.rand(50)
y2 = 5 * np.random.rand(50) + 25
y3 = 30 * np.random.rand(25)
y = np.concatenate([y1,y2,y3])

color_array = ['b']*50 + ['g']*50 + ['r']*25
plt.scatter(x, y, marker='D', c=color_array, s=[70])
plt.show()
```

## HeatMap

```python
data = np.random.randint(low=1, high=100, size=(10,10))
print(data)
sns.heatmap(data=data)
plt.show()
```

```python
pip install wordcloud
```

```python
import requests
from wordcloud import WordCloud, STOPWORDS
url = 'https://www.gutenberg.org/files/11/11-0.txt'
response = requests.get(url)
text = response.text
print(text)
```

```python
wc = WordCloud(background_color='black', max_words=2000)
wc.generate(text)
plt.imshow(wc)
plt.axis('off')
plt.show()

# size resembles frequency
# jetli vadhare size etli vadhare frequency
```

```python
wc = WordCloud(background_color='black', max_words=2000, stopwords=STOPWORDS)
# STOPWORDS :commonly used words ni list
# commonly used words remove thase
wc.generate(text)
plt.imshow(wc)
plt.axis('off')
plt.show()
```

## Regression Plot

```python
# Regression plot: Line + Scatter Plot
```

```python
x=[1,2,3,4,5,6]
y=[3,5,7,8,10,24]

sns.regplot(x=x, y=y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[3,5,7,8,10,24]

sns.regplot(x=x, y=y, scatter_kws={'s':10, 'alpha':0.7}, line_kws={'color':'red', 'linewidth':2})
plt.show()
```
