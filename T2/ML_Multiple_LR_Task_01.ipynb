```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/kavit88/Linear-Regression/main/CAR%20DETAILS%20FROM%20CAR%20DEKHO.csv')
```

```python
df.head(3)
```

```python
# aa data ma null or duplicate values nthi
# year categorical data che so no removal of outliers
# km_driven numerical data che but no removal of outliers as amuk ma 10000 hase or amuk ma 1000000 hase. so nthi jarur remove karvani
# so aa data ma data cleaning jarur nthi
```

```python
# name e Brand name -- eg. Maruti 800 AC : Maruti
# for i in range(len(df['name'])):
#     df['name'][i] = df['name'][i].split()[0]
# df.head(3)
```

```python
df['age'] = 2025-df['year']
df.drop(['year'], axis=1, inplace=True)
# year ni jagya e age naakhi etle year dropped
# model training ma name ni jarur nthi etle name dropped

# categorical ne numerical ma convert krva mate
df = pd.get_dummies(df, drop_first=True) # new version ma 0 or 1 aavse and old version ma False or True aavse
# model ne train karava ma minimum cols aapva mate aapde drop first karayu
# like 5 col hoy and drop first karaiye toh 4 thai jase 
# and all 4 ma 0 hase value toh automatic drop kareli column no data che em interpret thai jase
# drop_first alphabetical order ma first col remove karse  
df.head(3)
```

```python
# df.corr()
```

```python
x=df.drop('selling_price', axis=1)
y=df['selling_price']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)

lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
```

```python
print('Sq Err:',mean_squared_error(y_test, y_pred))
print('Abs Err:',mean_absolute_error(y_test, y_pred))
print('R2 Score:',r2_score(y_test, y_pred))
```
