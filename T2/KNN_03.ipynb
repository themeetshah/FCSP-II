```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/kavit88/Linear-Regression/main/Churn.csv')
df.head()
```

```python
df.drop(['Unnamed: 0', 'customerID'], axis=1, inplace=True)
df.dropna(inplace=True)
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
x = df.drop('Churn_Yes', axis=1)
y = df['Churn_Yes']

s = StandardScaler()
x = s.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cs = confusion_matrix(y_test, y_pred)
print(cs)
```

```python
#                 Actual Values
#                   No   Yes
# Predicted   No    TN   FN
# Values     Yes    FP   TP

# Actual Yes and Predicted No : FN (False Negative)
# [[937, 96]
#  [230, 144]]

# accuracy = (TP+TN)/Total_Samples

# Error rate = 1 - accuracy
```

```python
acc = (cs[0][0]+cs[1][1])/(cs[0][0]+cs[1][1]+cs[0][1]+cs[1][0])
acc
```

```python
# accuracy = (TP+TN)/Total_Samples

# Error rate = 1 - accuracy

# Precision = TP/(TP+FP)   (Predicted Positive ma sachhi ketli hati)

# Sensitivity (Recall) = TP/(TP+FN)   (total positive mathi ketli sachhi positive Predict kari)

# Specificity = TN/(TN+FP)   (total negative mathi ketli sachhi negative Predict kari)
```

```python
Y_act = pd.DataFrame(['Cat', 'Cat', 'Cat', 'Cat', 'Cat', 'Dog', 'Dog', 'Dog', 'Dog', 'Dog', 'Cat', 'Dog'])
Y_pred = pd.DataFrame(['Cat', 'Dog', 'Cat', 'Cat', 'Cat', 'Dog', 'Cat', 'Cat', 'Dog', 'Dog', 'Dog', 'Cat'])

# by default jupyter alphabetical order ma je 1st aavse ene -ve lese
from sklearn.metrics import confusion_matrix
cs = confusion_matrix(Y_act, Y_pred)
print(cs)

Total_Samples = len(Y_pred)
TP = cs[1][1]
TN = cs[0][0]
FP = cs[0][1]
FN = cs[1][0]

accuracy = (TP+TN)/Total_Samples
error_rate = 1 - accuracy
precision = TP/(TP+FP)
sensitivity = TP/(TP+FN)
specificity = TN/(TN+FP)

print('accuracy:',accuracy)
print('error_rate:',error_rate)
print('precision:',precision)
print('sensitivity:',sensitivity)
print('specificity:',specificity)
```
