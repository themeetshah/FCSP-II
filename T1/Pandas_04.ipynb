## fill the null data cells

```python
df = pd.DataFrame({'A':[1,2,np.nan,4,np.nan],
                   'B':[np.nan,np.nan,7,np.nan,9],
                   'C':[10,11,np.nan,np.nan,14]})
df
```

```python
df1 = df.fillna(method='ffill')  # forward fill
df1
```

```python
df2 = df.fillna(method='bfill')  # backward fill
df2
```

```python
# We'll use this method to fill null values
df3 = df.fillna({'A':50,'B':70,'C':90})  # A col ma je null hoy tya 50, B ma 70, C ma 90
df3
```

```python
df['A'].fillna(df['A'].mean())  # mean fill karse
```

```python
df['A'].fillna(df['A'].mean(), inplace=True)  # mean fill karse
# inplace=True lakhine print karaisu toh None print karse
df
```

```python
df['B'].fillna(df['B'].median(), inplace=True)  # median fill karse
df
```

```python
df['C'].fillna(df['C'].mode()[0], inplace=True)  # mode fill karse
# need to write [0] as mode may return more than one value
df
```

```python
#  Using ‘supermarket_sales.csv’ file do the following operations and give required answer by using proper programming process.
#  1).  Load the dataset into a pandas DataFrame and read first 8 rows.
#  2). Check for missing values and fill it by mean values of that particular column if any.
#  3).  Find the number of orders which have ‘Quantity’ less than 3 and which have (either ‘Rating’ greater than 8.5 or ‘Total’ 
# greater than 600).
#  4). Find the sum of ‘Total’ purchasing price spent by Member and Normal 'Customer type'.
#  5). Find the percentage of total of ‘gross income’ based on the different ‘Payment’ methods used by customers. (Ewallet, Cash 
# and Credit card)
#  6). Analyze the purchasing behavior of male and female customers using ‘Gender’ column. Find their average purchase prices 
# using ‘Total’ column.
#  7). Create a scatter plot that shows the relationship between total amount spent and rating. (keep ‘+’ marker, with marker size 
# 100 and green color).
#  8). Create a box plot that shows the distribution of ‘Rating’ and ‘Quantity’. And comment about outliers in both columns.
#  9). Visualize with parallel co-ordinates for ‘Unit price’, ‘Total’, ’cogs’ columns’ data with respect to ‘Product line’
```

```python
supermarket = pd.read_csv('supermarket_sales.csv')
df = pd.DataFrame(supermarket)
df.head(8)
```

```python
df.isna().sum()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
df.isna().sum()
```

```python
df[(df['Quantity']<3) & ((df['Rating']>8.5) | (df['Total']>600))]
```

```python
member = df[(df['Customer type']=='Member')]['Total'].sum()
normal = df[(df['Customer type']=='Normal')]['Total'].sum()
print("Normal:",normal)
print("Members:",member)
```

```python
total_gross = df['gross income'].sum()
ewallet = df[df['Payment']=='Ewallet']['gross income'].sum()
cc = df[df['Payment']=='Credit card']['gross income'].sum()
cash = df[df['Payment']=='Cash']['gross income'].sum()
print("Ewallet %:",(ewallet*100/total_gross))
print("Credit Card %:",(cc*100/total_gross))
print("Cash %:",(cash*100/total_gross))
```

```python
male = df[df['Gender']=='Male']['Total'].mean()
female = df[df['Gender']=='Female']['Total'].mean()
print("avg. purchase price by Male:",male)
print("avg. purchase price by Female:",female)
```

```python
total = list(df['Total'])
rating = list(df['Rating'])
plt.scatter(total,rating,marker='+',s=100,color='green')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.boxplot([df['Rating'],df['Quantity']])
plt.show()
```

```python
p = pd.plotting.parallel_coordinates(df,'Product line',cols=['Unit price','Total','cogs'])
plt.show()
```
